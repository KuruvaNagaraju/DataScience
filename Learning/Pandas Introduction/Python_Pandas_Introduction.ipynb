### Create a dataframe with a list as input

```python
## Create a list which contains list of lists.
stduentData = [['Alex',10],['Bob',12],['Clark',13]]
stduentData
```

```python
## Create a dataframe with list and assign column names as Names,Age and display it.
sdata=pd.DataFrame(stduentData,columns=['Names','Age'])
sdata
```

```python
## Create a dataframe with two lists.
student=pd.DataFrame([10,12,13],['Alex','Bob','Clark'])
student
```

## **loc (to subset data by Row/Column labels)**

```python
## Label based indexing make more sense for column subsetting.
## Get first 3 records of ipl_final data and store in a varible and display it.
subset1 =ipl_final.head(3)
subset1
```

```python
subset1=ipl_final.loc[:,['venue','winner']] ## It will return all rows and venue,winnner columns
subset1.head() ## it will return first records 
```

```python
## For loc based subsetting using row and column indexer together is rare.
subset2 =ipl_new.head(3)
subset2
```

```python
subset2=ipl_final.loc[['2009_1','2010_1','2011_1'],['venue','winner']] ## it will returrn 2009_1,2010_1,2011_1 rows and venue,winner columns data
subset2.head() ## It will return first 5 records by default but here we are mentioning 3 rows
```

## **iloc    (to subset data based on Row/Column number)**

```python
## Numeric based indexing usually makes sense for row subsetting 
```

```python
subset3=ipl_final.iloc[5:10,:] ## It will return 5,6,--9 rows and all columns data.
subset3.head() ## It will return 5 records from subset.
```

```python
subset4 = ipl_final.iloc[70:91,3:7] ## It will return 70-90 rows and 3-6 columns data.
subset4.head() ## It will return first 5 records.
```

## **Boolean indexing (to subset based on value in each cell of the column )**
**Here we subsetting the data where the cell value in "City" column is equal to "Hyderabad"**

```python
subset5 = ipl_final.loc[ipl_final['city']=='Hyderabad',:] ## it will return the records which city is equal to Hyderabad.
subset5.head(3)
```
