### Analyzing a single dataframe (and its optional target feature)

```python
## Analyze train data.
%time my_report = sweetviz.analyze([train, "Train"],target_feat='SalePrice')
```

```python
## Prepare HTML file with train data analysis.
%time my_report.show_html('Report.html')
```

### Comparing two dataframes (e.g. Test vs Training sets)

```python
## Compare train data with test data.
%time my_report1 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```

```python
## Prepare HTML file with analysis.
%time my_report1.show_html('Report_1.html')
```
